```python
# @title INITIALIZE FUNCTIONS - RUN ON (RE)START ONLY

#log function that will push out every action on the MQTT server (ONLY FOR DEBUG)
def on_log(client, userdata, level, buf):
  print( str(datetime.datetime.now()) + ": ",buf)


#subscribe handler
def on_message(client, userdata, message):

  #pull in the global data frame that is being concat
  global df_data

  #print ("Client received: " + str(message.topic))       #(ONLY FOR DEBUG)
  #print ("Message received: " + str(message.payload))    #(ONLY FOR DEBUG)

  df = pd.DataFrame(json.loads(str(message.payload, encoding='utf-8')))
  df = df.reindex(columns=['time', 'data', 'ambient','sensor']).sort_values(by=['time'],ignore_index=True)

  df_data = pd.concat([df_data, df], ignore_index=True)
  #print(df_data.tail(20))                                #(ONLY FOR DEBUG)



#connection handler
def on_connect(client, userdata, flags, rc):
 if rc == 0:
  print("Connected to broker")
  global Connected
  Connected = True
 else:
  print("Connection failed")


print('Everything loaded ok')
```

```python
# @title A: Setup and Connect MQTT Client for Streaming


Connected = False
# using public available service from hivemq
broker_address="test.mosquitto.org"
port = 1883
# user = "roger"            # Only for Secured MQTT Broker
# password = "password"     # Only for Secured MQTT Broker

df_data = pd.DataFrame({'time':[],'data':[],'ambient':[],'sensor':[]}, columns=['time', 'data', 'ambient','sensor'])

client = mqttClient.Client("ColabViz")  #create new instance
client.on_connect = on_connect          #attach function(s) to callback
client.on_message = on_message
#client.on_log = on_log                             #(ONLY FOR DEBUG)
#client.username_pw_set(user, password=password)    # Only for Secured MQTT Broker
client.connect(broker_address, port=port)

#fire up the MQTT monitoring loop but nothing in it until subscriptions are made
client.loop_start()

print('Dataframe and MQTT Broker all ready to go')
```

```python
# @title B: Get data and show streaming plot

#Make sure the connection to broker is alive
while Connected != True:
  time.sleep(0.1)

#create subscriptions to broker topics
client.subscribe("test/results")    #Sensor data in JSON packets
#client.subscribe("test/info")      #Not implemented at this time; study header


try:
  print('Starting capture....')

  while True:
    time.sleep(1)
    clear_output(wait=True)
    df_data.tail(500).plot(x='time',y='data',figsize=(15,5))
    plt.ylim(10, 250)
    plt.ylabel("Temperature [C]")
    plt.xlabel("Time from Restart [ms]")
    plt.show()

except KeyboardInterrupt:
 client.disconnect()
 client.loop_stop()
 print('Exited gracefully')
```

```python
# @title C: Plot a Summary of the Data

output_notebook()

graph = figure(title="Summary of the Capture")

graph.line(source=df_data,x='time',y='data', color='red',legend_label='Thermocouple')
graph.line(source=df_data,x='time',y='ambient', color='blue',legend_label='Ambient')
graph.y_range = Range1d(10, 250)
graph.yaxis.axis_label = 'Temperature [C]'
graph.xaxis.axis_label = 'Time from Reset [ms]'
graph.height=600
graph.width=1200
show(graph)

#Sources:
#https://docs.bokeh.org/en/3.0.2/docs/user_guide/basic/axes.html
#https://programminghistorian.org/en/lessons/visualizing-with-bokeh
```
